```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mnist_train.csv', header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:, 1:]
X
```

```python
Y = df.iloc[:, 0]
Y
```

```python
Y.value_counts()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
X_train.iloc[0]
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
ppn = Perceptron()
ppn.fit(X_train, Y_train)
```

```python
y_pred = ppn.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print(accuracy)
```

```python
my_handwritten_digit = pd.read_csv('numbers.csv', header=None)
my_handwritten_digit
```

```python
y_pred_handwrite = ppn.predict(my_handwritten_digit)
y_pred_handwrite
```

```python
my_handwrite_correctly = accuracy_score(y_pred_handwrite, my_handwritten_digit.index)
my_handwrite_correctly
```

```python
lR = LogisticRegression(max_iter=100)
lR.fit(X_train, Y_train)
```

```python
accuracy_lr = accuracy_score(Y_test, lR.predict(X_test))
accuracy_lr
```

```python
accuracy_lr_handwrite = accuracy_score(y_pred_handwrite, my_handwritten_digit.index)
accuracy_lr_handwrite
```
